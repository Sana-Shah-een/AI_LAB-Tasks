```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from datasets import load_dataset
```

```python
data=load_dataset("tnpb/breas-cancer-wisconsin-kaggle")
```

```python
df = pd.DataFrame(data['train'])
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum().sum())
```

```python
print(df.dtypes)
```

```python
df.describe()
```

```python
radius_mean = df['radius_mean'].values
radius_mean_mean = np.mean(radius_mean)
radius_mean_median = np.median(radius_mean)
radius_mean_variance = np.var(radius_mean)
radius_mean_std = np.std(radius_mean)

# Print the numpy statistical calculations
print(f"Radius Mean - Mean: {radius_mean_mean}")
print(f"Radius Mean - Median: {radius_mean_median}")
print(f"Radius Mean - Variance: {radius_mean_variance}")
print(f"Radius Mean - Standard Deviation: {radius_mean_std}")
```

```python
import seaborn as sns
diagnosis_counts = df['diagnosis'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(diagnosis_counts)))
plt.title('Distribution of Diagnoses')
plt.show();
```
