```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from datasets import load_dataset
```

```python
data=load_dataset("maharshipandya/spotify-tracks-dataset")
```

```python
df = pd.DataFrame(data['train'])  # Or the specific split you want
df.head()
```

```python
df.sort_values(by='popularity', ascending=True)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df.fillna(0, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
df.describe()
```

```python
# Artist frequency
artist_counts = df['artists'].value_counts().head(10)
artist_counts
```

```python
plt.hist(df['popularity'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
genre_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)
print(genre_popularity)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(genre_popularity.index, genre_popularity.values, color='orange')
plt.title('Top 10 Genres by Average Popularity')
plt.xlabel('Track Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
top_tracks = df[['track_name', 'artists', 'popularity']].sort_values(by='popularity', ascending=False).head(10)
top_tracks
```

```python
# Plot top 10 tracks by popularity
plt.figure(figsize=(10, 6))
plt.barh(top_tracks['track_name'], top_tracks['popularity'], color='lightblue')
plt.title('Top 10 Most Popular Tracks')
plt.xlabel('Popularity')
plt.ylabel('Track Name')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
# Select relevant numerical columns
numerical_columns = ['popularity', 'danceability', 'energy', 'loudness', 'tempo', 'duration_ms']

# Convert these columns into a NumPy array
numerical_data = df[numerical_columns].to_numpy()

numerical_data[:5]  # Display first 5 rows of the array
```

```python
mean_values = np.mean(numerical_data, axis=0)
print("Mean values of numerical columns:", mean_values)
```

```python
median_values = np.median(numerical_data, axis=0)
print("Median values of numerical columns:", median_values)
```

```python
std_values = np.std(numerical_data, axis=0)
print("Standard deviation of numerical columns:", std_values)
```

```python
total_duration = np.sum(numerical_data[:, numerical_columns.index('duration_ms')])
print(f"Total duration of all tracks: {total_duration} ms")
```

```python
min_values = np.min(numerical_data, axis=0)
max_values = np.max(numerical_data, axis=0)

print("Min and Max values for numerical columns:")
for col, min_val, max_val in zip(numerical_columns, min_values, max_values):
    print(f'{col}: Min = {min_val}, Max = {max_val}')
```

```python
# Plot the distribution of top 10 artists
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(artist_counts.index, artist_counts.values, color='purple')
plt.title('Top 10 Artists by Number of Tracks')
plt.xlabel('Artist Name')
plt.ylabel('Track Count')
plt.xticks(rotation=45)
plt.show()
```
